```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = 'hr_employee_churn_data.csv'
df = pd.read_csv(path)
df.head()
```

```python
# @title salary

df.groupby('salary').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)

# It seems the employees with medium and low salaries are more than those with high salaries
```

```python
# @title Satisfaction Level by Salary

salaries = df['salary'].unique()

data = []
for salary in salaries:
    data.append(df[df['salary'] == salary]['satisfaction_level'])

plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=salaries)
plt.xlabel('Salary')
plt.ylabel('Satisfaction Level')
_ = plt.title('Satisfaction Level by Salary')
```

```python
# @title last_evaluation

df['last_evaluation'].plot(kind='hist', bins=20, title='last_evaluation')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title satisfaction_level

df['satisfaction_level'].plot(kind='hist', bins=20, title='satisfaction_level')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title empid

df['empid'].plot(kind='hist', bins=20, title='empid')
plt.gca().spines[['top', 'right',]].set_visible(False)
```
